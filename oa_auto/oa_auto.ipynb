```python
import turtle as tt
def drawCircleBasic(x,y,r=2,color='white'):
    tt.pu()  #抬起画笔
    tt.goto(x,y)  #起始位置
    tt.pd()   #putdown放下画笔
    tt.color(color)
    tt.circle(r, 360)  #半径，角度

tt.pensize(3)  #画笔尺寸设置为3

drawCircleBasic(0,-120)
tt.hideturtle()
tt.done()
```

turtle绘制雪花❄️

```python
import turtle as tt
import random
def snowleaf():
    x=random.randint(-350, 350)
    y=random.randint(-270, 270)
    size=random.randint(1,30)

    tt.pu()  #put up抬起画笔
    tt.goto(x,y)
    tt.pd()
    # tt.color('lightblue')
    tt.color('white')
    dens=random.randint(10,11)  #设置边数
    for _ in range(dens):
        tt.goto(x,y)
        tt.forward(size)  #当前画笔方向向前移动
        tt.backward(size /3)
        tt.right(15)
        tt.forward(size /2)
        tt.backward(size /2)
        tt.right(30)
        tt.forward(size /2)

        tt.backward(size /2)
        tt.right(15)
        tt.forward(size / 3)
        tt.right(180-(360 / 9))  #顺时针旋转360 / dens度

def drawCircle(color='white'):
    x=random.randint(-350, 350)
    y=random.randint(-270, 270)
    r=random.randint(1,10)
    tt.pu()  #抬起画笔
    tt.goto(x,y)  #起始位置
    tt.pd()   #putdown放下画笔
    tt.begin_fill()
    tt.color(color)
    tt.circle(r, 360)  #半径，角度
    tt.end_fill()


tt.pensize(3)
tt.bgcolor('black')

count=30
for i in range(count):
    snowleaf()
for i in range(3*count):
    drawCircle()

tt.hideturtle()
tt.done()
```

樱花树

```python
from turtle import *
from random import *
from math import *

def tree(n,l):
    pd()#下笔
    #阴影效果
    t = cos(radians(heading()+45))/8+0.25
    pencolor(t,t,t)
    pensize(n/3)
    forward(l)#画树枝

    if n>0:
        b = random()*15+10 #右分支偏转角度
        c = random()*15+10 #左分支偏转角度
        d = l*(random()*0.25+0.7) #下一个分支的长度
        #右转一定角度,画右分支
        right(b)
        tree(n-1,d)
        #左转一定角度，画左分支
        left(b+c)
        tree(n-1,d)
        #转回来
        right(c)
    else:
        #画叶子
        right(90)
        n=cos(radians(heading()-45))/4+0.5
        pencolor(n,n*0.8,n*0.8)
        circle(3)
        left(90)
        #添加0.3倍的飘落叶子
        if(random()>0.7):
            pu()
            #飘落
            t = heading()
            an = -40 +random()*40
            setheading(an)
            dis = int(800*random()*0.5 + 400*random()*0.3 + 200*random()*0.2)
            forward(dis)
            setheading(t)
            #画叶子
            pd()
            right(90)
            n = cos(radians(heading()-45))/4+0.5
            pencolor(n*0.5+0.5,0.4+n*0.4,0.4+n*0.4)
            circle(2)
            left(90)
            pu()
            #返回
            t=heading()
            setheading(an)
            backward(dis)
            setheading(t)
    pu()
    backward(l)#退回

bgcolor(0.5,0.5,0.5)#背景色
ht()#隐藏turtle
speed(0)#速度 1-10渐进，0 最快
tracer(0,0)
pu()#抬笔
backward(100)
left(90)#左转90度
pu()#抬笔
backward(300)#后退300
tree(12,100)#递归7层
done()
```
